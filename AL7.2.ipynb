## 알고리즘 예제 7.2


1. 문제정의: 

피보나치수열(테이블화 이용)


2. 알코리즘 성명:
- `f = [None] * (n+1)` n+1 크기의 리스트(테이블)을 만든다.
- `f[0] = 0 , f[1] = 1` f에 0과 1을 저장
- `for i in range(2, n+ 1):` 반복을을 통해 `f[i]`를 증가 시키며 상향식으로 피보나치 수열 값을 테이블에 저장한다.

3. 손으로 푼 예제 (이미지삽입):





4. 코드 개요(함수, 입력변수, 출력변수):


- 함수 : 
    - `def fib_dp_tab(n)` 피보나치 수열을 테이블화(상향식)를 이용한 동적 계획법

- 입력 변수 :  
    - `n` 피보나치 수열을 계산하는 항

- 출력 변수 : 
    - `f[n]` 피보나치 수열의 n번째 값


```python
# 5. 코드
import time

def fib_dp_tab(n) :
    f = [None] * (n+1)
    f[0] = 0 
    f[1] = 1
    for i in range(2, n+ 1):
        f[i] = f[i-1] + f[i-2]
    return f[n]
```

```python
# 6. 테스트 코드 
n = 10000
mem = [None] * (n + 1)


start = time.time()
result = fib_dp_tab(n)
end= time.time()

time = end - start

print('동적계획(테이블화): Fibonacci(%d) = %d' % (n, result))
print('실행 시간: %.5f 초' % time)
```
