## 알고리즘 예제 7.1

1. 문제정의: 

피보나치수열(메모이제이션 이용)

2. 알코리즘 성명:

- `if ( mem[n] ==  None) :` 아직 풀리지 않은 경우인지 검사
- `if n < 2:` 6행~9행 아직 풀리지 않았다면 계산하고 저장
- `return mem[n]` 이미 풀린 경우나 풀리지 않은 경우를 계산한 결과를 `mem[n]`을 반환
    
피보나치 수열을 계산하는 함수로 메모이제이션을 이용하는데 이미 계산된 피보나치 수를 메모리에 저장하고 다른 수열을 계산할 때 저장된 값을 사용하여 재활용한다. 


3. 손으로 푼 예제 (이미지삽입):





4. 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def fib_dp_mem(n)` 피보나치 수열을 메모이제이션(하향식)을 이용한 동적 계획법

- 입력 변수 :  
    - `n` 피보나치 수열을 계산하는 항

- 출력 변수 : 
    - `mem[n]` 피보나치 수열의 n번째 값

```python
# 5. 코드
import time

def fib_dp_mem(n):
    if ( mem[n] ==  None) :
        if n < 2:
            mem[n] = n
        else:
            mem[n] = fib_dp_mem(n-1) + fib_dp_mem(n-2)
    return mem[n]
```

```python
# 6. 테스트 코드 
n = 2500
mem = [None] * (n + 1)


start= time.time()
result = fib_dp_mem(n)
end = time.time()

time = end - start

print('동적계획(메모이제이션): Fibonacci(%d) = %d' % (n, result))
print('실행 시간: %.5f 초' % time)
```
